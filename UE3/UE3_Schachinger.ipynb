```python
db = client.get_database('sample_mflix')
movies = db.movies.find()
movies_df = pd.DataFrame(movies)
movies_df.head()


```

```python
movies_df[movies_df.loc[:,'title'].str.contains('.\s.') == False].shape[0]

```

Pick all movies that match following criteria: 

* imdb.rating is at least 7
* genres does not contain "Crime" or "Horror"
* rated is either "PG" or "G"


```python

rate_bool =[]
for i in [c['rating'] for c in [x for x in movies_df.loc[:,'imdb']]]:
    if i == '':
        i=0
    rate_bool.append(i>7)

genre_bool = []
for movie in [mov for mov in movies_df.loc[:,'genres']]:
    if type(movie) == float:
        movie =[]
    genre_bool.append(('Horror' in movie) | ('Crime' in movie))


rated_bool = [(m in ['PG','G']) for m in movies_df.loc[:,'rated']]
```

```python
movies_df[np.multiply(np.multiply(rate_bool,genre_bool),rated_bool)]

```
